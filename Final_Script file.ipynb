# Plotting pie chart for Division name,Department name and Class name:

```python
#Create wedges and corresponding labels for column in dataframe.
def NameCount (insidedataframe, column):
    #Store unique entries 
    names = insidedataframe[column].unique()
    #remove nans
    names = [x for x in names if str(x) != 'nan']
    #Create array to store values
    count = np.empty(len(names))
    #Store recurrence of each value
    for i in range(len(names)):
        count[i] = p[(p[column] == names[i])].shape[0]
    return names, count
```

```python
#Division
div_name, div_count = NameCount(p, 'Division Name')
#Department 
dep_name, dep_count = NameCount(p, 'Department Name')
#Class
class_name, class_count = NameCount(p, 'Class Name')
```

```python
plt.pie(div_count, labels=div_name, autopct='%1.2f%%', shadow=True, radius= 1)
plt.title("Division", fontsize = 20)
plt.show()
```

```python
plt.pie(dep_count, labels=dep_name, autopct='%1.1f%%', shadow=True, radius= 1)
plt.title("Department", fontsize = 20)
plt.show()
```

```python
plt.pie(class_count, labels=class_name, autopct='%1.1f%%', shadow=True, radius= 1)
plt.title("Class", fontsize = 20)
plt.show()
```

```python
p= p.dropna()
reviews=p['Review Text'].tolist()
recommend=p['Recommended IND'].tolist()
assert len(reviews) == len(recommend)
```

# Most Common Words

```python
#create three counter objects to store positive, negative, and total counts
positive_counts = Counter()
negative_counts = Counter()
total_counts = Counter()
```

```python
#loop over all the words in all the reviews and increment the count
for i in range(len(reviews)):
    if recommend[i]==1:
        for word in reviews[i].split(" "):
            positive_counts[word] +=1
            total_counts[word] +=1
    else:
        for word in reviews[i].split(" "):
            negative_counts[word] += 1
            total_counts[word] += 1
```

### Most common words that were in positive feedbacks

```python
positive_counts.most_common()[:10]
```

### Most common words that were in negative feedbacks

```python
negative_counts.most_common()[:12]
```

# Sentiment Analysis:

```python
rvt=np.array(cp['Review Text'])
```

```python
positive_words=['love','good','great','amazing','super','smooth','fits','nice','perfect','like']
```

```python
negative_words=['hate','horrible','big','small','rough','ugly','expensive','damage','never']
```

```python
sentiment=np.zeros(len(rvt))
```

```python
import re      #Regular Expression
for j in range(len(positive_words)):
    for k in range(len(rvt)):
        rv=re.sub("[^\w]"," ",str(rvt[k])).split()
        for p in range(len(rv)):
            if positive_words[j]==rv[p]:
                sentiment[p] += 1
            
```

### This is how many Number of positive words which were in review text

```python
np.sum(sentiment)
```

```python
import re
for j in range(len(negative_words)):
    for k in range(len(rvt)):
        rv=re.sub("[^\w]"," ",str(rvt[k])).split()
        for p in range(len(rv)):
            if negative_words[j]==rv[p]:
                sentiment[p] += 1
            
```

### This is how many Number of negative words which were in review text

```python
np.sum(sentiment)
```
